```python
rows = []
for col in df.columns:
    dtype   = str(df[col].dtype)
    nulls   = df[col].isnull().sum()
    pct_null = f"{nulls / len(df) * 100:.1f}%"
    nuniq   = df[col].nunique()
    sample  = df[col].dropna().iloc[0] if nulls < len(df) else ""
    grupo, desc = COLUMN_META.get(col, ("Otros", ""))
    rows.append({
        "Grupo":         grupo,
        "Columna":       col,
        "Tipo":          dtype,
        "Únicos":        nuniq,
        "Nulos":         nulls,
        "% Nulo":        pct_null,
        "Ejemplo":       str(sample),
        "Descripción":   desc,
    })

schema = pd.DataFrame(rows)
```

```python
GROUP_COLORS = {
    "Identificadores":          "#dbeafe",
    "Datos personales":         "#fef9c3",
    "Comisiones":               "#dcfce7",
    "Trayectoria legislativa":  "#fce7f3",
    "Trayectoria administrativa":"#ede9fe",
    "Trayectoria juvenil":      "#ffedd5",
    "Formación académica":      "#cffafe",
    "Conteos de trayectoria":   "#f1f5f9",
    "Otros":                    "#ffffff",
}

def row_color(row):
    color = GROUP_COLORS.get(row["Grupo"], "#ffffff")
    return [f"background-color: {color}" for _ in row]

styled = (
    schema.style
    .apply(row_color, axis=1)
    .set_properties(**{"text-align": "left", "font-size": "13px"})
    .set_table_styles([
        {"selector": "th", "props": [
            ("background-color", "#1e293b"),
            ("color", "white"),
            ("font-size", "13px"),
            ("text-align", "left"),
            ("padding", "6px 10px"),
        ]},
        {"selector": "td", "props": [
            ("padding", "4px 10px"),
            ("border-bottom", "1px solid #e2e8f0"),
        ]},
        {"selector": "", "props": [
            ("border-collapse", "collapse"),
            ("width", "100%"),
        ]},
    ])
    .hide(axis="index")
)

display(styled)
```

## Summary by group

```python
summary = (
    schema.groupby("Grupo", sort=False)
    .agg(n_columnas=("Columna", "count"), total_nulos=("Nulos", "sum"))
    .reset_index()
)

display(
    summary.style
    .set_properties(**{"text-align": "left", "font-size": "13px"})
    .set_table_styles([{"selector": "th", "props": [
        ("background-color", "#1e293b"), ("color", "white"),
        ("padding", "6px 10px"), ("text-align", "left")
    ]}])
    .hide(axis="index")
)
```
